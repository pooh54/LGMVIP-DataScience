```python
import cv2
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display

```

```python
import numpy as np
```

```python
image = cv2.imread('/content/peacock.jpg')
big = cv2.resize(image, (500, 500))
cv2_imshow(cv.cvtColor(big, cv.COLOR_BGR2RGB))
```

```python
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
bigger = cv2.resize(gray_image, (500,500))
cv2_imshow(bigger)
```

```python
neg_img = cv2.bitwise_not(image)
big_img= cv2.resize(neg_img, (500,500))
cv2_imshow(big_img)
```

```python
blurred = cv2.GaussianBlur(gray_image, (21, 21), 0)
bigger_img= cv2.resize(blurred, (500,500))
cv2_imshow(bigger_img)
```

```python
inverted_blurred = 255 - blurred
```

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
bigger = cv2.resize(pencil_sketch, (500, 500))
cv2_imshow(bigger)
```
